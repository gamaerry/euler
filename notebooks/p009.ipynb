A Pythagorean Triplet is a set of three natral numbers a<b<c for wich

a²+b²=c²

There exists exactly one Pythagorean triplet for wich a+b+c=1_000. Find the product abc

```python
n = 1_000
```

# Primer intento:

Primer enfoque basado en recorrer exhaustivamente todos los posibles triples crecientes y comprobar tanto la suma como la condicion pitagorica.

```python
def primer_intento():
    resultado = 0
    encontrado = False
    for i in range(1, n):
        for j in range(i + 1, n):
            for k in range(j + 1, n):
                if i + j + k == n and i**2 + j**2 == k**2:
                    resultado = i * j * k
                    encontrado = True
                    break
            if encontrado:
                break
        if encontrado:
            break
    return resultado

primer_intento()
```

# Optimizacion:

chatgpt recomienda usar la restriccion a+b+c=n para eliminar una variable y despejar b en funcion de a, reduciendo la busqueda a una sola iteracion con aritmetica entera:

```python
def optimizacion():
    for a in range(1, n // 3):
        numerador = n * (n - 2 * a)
        denominador = 2 * (n - a)
        if numerador % denominador == 0:
            b = numerador // denominador
            return a * b * (n - a - b)
    return 0

optimizacion()
```
